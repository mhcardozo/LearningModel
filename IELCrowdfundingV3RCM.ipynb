```python
import os
import numpy as np
import pandas as pd
import csv
import random
from matplotlib import pyplot as plt
from matplotlib import colors as mcolors

# random.seed(42)
# np.random.seed(42)
random.seed(12345)
np.random.seed(12345)

class Agent:
    def __init__(self, agent_type, possible_actions, action_set, exper_rate, lambda_rate, threshold, memory, always_choose_43=False, always_choose_13=False):
        self.agent_type = agent_type
        self.action_set = action_set
        self.possible_actions = possible_actions
        self.exper_rate = exper_rate
        self.lambda_rate = lambda_rate
        self.threshold = threshold
        self.always_choose_43 = always_choose_43
        self.always_choose_13 = always_choose_13
        self.action_count = [1, 1, 1]
        self.probabilities = np.array([1/3, 1/3, 1/3])
        self.average_fitness = np.array([80, 80, 80])
        self.actions_chosen_record = []  # Record of actions chosen
        self.generation = 1  # Start from generation 1, will be updated in the main loop
        self.memory = memory # periods agents remember
        self.fitness_history = {action: [80] * self.memory for action in action_set}
        
    def update_action_set(self):
        temp_actions = []
        for _ in range(3):
            choices = np.random.choice(self.action_set, 2, replace=True)
            indices = [self.action_set.index(choice) for choice in choices]
            selected_index = indices[0] if self.average_fitness[indices[0]] >= self.average_fitness[indices[1]] else indices[1]
            temp_actions.append(self.action_set[selected_index])

        self.action_set = temp_actions
        
        # Experimentation phase
        for i, actions in enumerate(self.action_set):
            if np.random.rand() < self.exper_rate:
                #temp_actions0 = []
                choices0 = np.random.choice(self.possible_actions, 1, replace=True)
#                 indices0 = [self.possible_actions.index(choice) for choice in choices0]
#                 selected_index0 = indices[0]
                self.action_set[i] = choices0[0]

    def choose_action(self):
        if self.always_choose_43:
            action_chosen = 43
        elif self.always_choose_13:
            action_chosen = 13
        else:
            action_chosen = np.random.choice(self.action_set, p=self.probabilities) 
    
        self.actions_chosen_record.append(action_chosen)  # Update the record
        return action_chosen
        
    def calculate_fitness(self, group_contribution, action_chosen):
        #self.action_count[self.action_set.index(action_chosen)] += 1
        factor = 0.4 if self.agent_type == 'high' else 0.2

        if group_contribution < self.threshold:
            fitness_chosen = 80
        else:
            fitness_chosen = factor * group_contribution + 16 + 80 - action_chosen if action_chosen == 43 else factor * group_contribution + 80 - action_chosen
            print(f'Agent type: {self.agent_type}, action chosen: {action_chosen}, group_contribution: {group_contribution}, fitness_chosen: {fitness_chosen}')
        # for i, action in enumerate(self.action_set):
        #     if action != action_chosen:
        #         fitness_not_chosen = [factor * (group_contribution - action_chosen + action) + (
        #             16 if action == 43 else 0) + 80 - action if group_contribution - action_chosen + action >= self.threshold else 80
        #                               for action in self.action_set if action != action_chosen]

        self.average_fitness[self.action_set.index(action_chosen)] = fitness_chosen

        for i, action in enumerate(self.action_set):
            if action != action_chosen:
                fitness_not_chosen = factor * (group_contribution - action_chosen + action) + (16 if action == 43 else 0) + 80 - action if group_contribution - action_chosen + action >= self.threshold else 80
                self.average_fitness[i] = fitness_not_chosen
                
                
#         # Update the fitness history for the chosen action
#         self.fitness_history[action_chosen].pop(0)  # Remove the oldest fitness
#         self.fitness_history[action_chosen].append(fitness_chosen)  # Add the current fitness to the end
#         # Calculate the average fitness for the chosen action
#         self.average_fitness[self.action_set.index(action_chosen)] = np.mean(self.fitness_history[action_chosen])
        
#         # Update the fitness history for the not chosen actions
#         for action in self.action_set:
#             if action != action_chosen:
#                 # Calculate fitness_not_chosen as before
#                 self.fitness_history[action].pop(0)  # Remove the oldest fitness
#                 self.fitness_history[action].append(fitness_not_chosen)  # Add the current fitness to the end
#                 # Calculate the average fitness for the not chosen action
#                 self.average_fitness[self.action_set.index(action)] = np.mean(self.fitness_history[action])

        self.update_probabilities()


    def update_probabilities(self):
        total_payoff = sum(self.average_fitness)
        #self.probabilities = [payoff/total_payoff for payoff in self.average_fitness]
        self.probabilities = np.exp(self.lambda_rate * self.average_fitness) / np.sum(np.exp(self.lambda_rate * self.average_fitness))
        
    def reset_actions_chosen_record(self):
        self.actions_chosen_record = []


############ Initialize parameters of the model:
n_simul = 50 #2

n_agents = 20 # Define the total number of agents
n_generations = 30
possible_actions = [0, 13, 43]
action_set = possible_actions
decay_rate = 0.15
exper_rate = np.exp(-decay_rate*1)#0.35/(1+(agent.generation-1)/5000)#
lambda_rate = 5
memory = n_generations
threshold = 70 # contribution theshold
high_agent_prob = 0.2


averages_high = np.zeros((n_generations, len(action_set)))
averages_low = np.zeros((n_generations, len(action_set)))

for simul in range(n_simul):
    print(f'Simulation: {simul}')
    # Generate agents
    agents = []
    n_high_agents = int(n_agents * high_agent_prob)  # 20% of agents are 'high' type
    n_high_agents_choose_43 = int(n_high_agents * 0)  # 10% of 'high' type agents always choose 43
    n_low_agents = n_agents - n_high_agents  # remaining 80% of agents are 'low' type
    n_low_agents_choose_13 = int(n_low_agents * 0)  # 10% of 'low' type agents always choose 13


    for _ in range(n_high_agents_choose_43):
        agents.append(Agent('high', possible_actions, action_set, exper_rate, lambda_rate, threshold, memory, always_choose_43=True))

    for _ in range(n_high_agents - n_high_agents_choose_43):
        agents.append(Agent('high', possible_actions, action_set, exper_rate, lambda_rate, threshold, memory))

    for _ in range(n_low_agents_choose_13):
        agents.append(Agent('low', possible_actions, action_set, exper_rate, lambda_rate, threshold, memory, always_choose_13=True))

    for _ in range(n_low_agents - n_low_agents_choose_13):
        agents.append(Agent('low', possible_actions, action_set, exper_rate, lambda_rate, threshold, memory))
    # Initialize record arrays for both agent types and each action
    high_agents_records_per_generation = np.zeros((n_generations, len(action_set)))
    low_agents_records_per_generation = np.zeros((n_generations, len(action_set)))

    for gen in range(n_generations):
        random.shuffle(agents)
        for i in range(0, n_agents, 4):
            group = agents[i:i+4]
            for agent in group:
                agent.update_action_set()
            group_contributions = [agent.choose_action() for agent in group]
            group_contribution = sum(group_contributions)
            for agent, action_chosen in zip(group, group_contributions):
                agent.calculate_fitness(group_contribution, action_chosen)


        # Update the generation for each agent after the fitness has been calculated
        for agent in agents:
            agent.generation += 1
            agent.exper_rate = np.exp(-decay_rate*agent.generation)#0.35/(1+(agent.generation-1)/5000)# # Experimentation rate becomes smaller with more generations

        if gen % 5 == 0:  # Print every 10 generations
            for agent in agents[:5]:
                print(f"Generation {gen}, Agent Type: {agent.agent_type}")
                print(f"Average Fitness: {agent.average_fitness}")
                print(f"Probabilities: {agent.probabilities}")
                #print(f"Action Counts: {agent.action_count}\n")
                print(f"Action Set: {agent.action_set}")

        # After each generation, update records_per_generation for each agent type
        for agent in agents:
            if agent.agent_type == 'high':
                high_agents_records_per_generation[gen] += np.bincount(agent.actions_chosen_record, minlength=max(action_set)+1)[action_set]/n_high_agents
            else:
                low_agents_records_per_generation[gen] += np.bincount(agent.actions_chosen_record, minlength=max(action_set)+1)[action_set]/n_low_agents

            agent.reset_actions_chosen_record()  # Reset actions_chosen_record for each agent

    averages_high += high_agents_records_per_generation / n_simul
    averages_low += low_agents_records_per_generation / n_simul

### Save series for graphing in R
averages_high_df = pd.DataFrame(averages_high, columns = ['Contract 1', 'Contract 2', 'Contract 3'])
averages_low_df = pd.DataFrame(averages_low, columns = ['Contract 1', 'Contract 2', 'Contract 3'])

averages = [averages_high_df,averages_low_df]

#folder_path = 'C:/Users/MarcO/Google Drive/Research/Crowdfunding PGoods/SESSIONS/Data4Analysis'
folder_path = 'G:/My Drive/Research/Crowdfunding PGoods/SESSIONS/Data4Analysis/'

# Loop through the DataFrames and save each to a CSV file
for i, df in enumerate(averages):
    if i == 0:
        file_name = 'averages_high_RCM.csv'
    else:
        file_name = 'averages_low_RCM.csv'

    # Define the full file path
    file_path = os.path.join(folder_path, file_name)

    df.to_csv(file_path, index=False)

    print(f"CSV file '{file_name}' has been saved in '{folder_path}'.")

    
# Plot
action_colors = {0: 'green', 13: 'red', 43: 'blue'}
action_labels = {0: 'Contract 1', 13: 'Contract 2', 43: 'Contract 3'}

# Plot for High Agents
plt.figure(figsize=(10, 5))
for i, action in enumerate(action_set):
    color = action_colors.get(action, 'black')
    label = action_labels.get(action, f'Action {action}')
    plt.plot(range(n_generations), averages_high[:, i], color=color, label=label)

plt.title('High Agents')
plt.xlabel('Period')
plt.ylabel('Choice Frequency')
plt.legend()
plt.show()

# Plot for Low Agents
plt.figure(figsize=(10, 5))
for i, action in enumerate(action_set):
    color = action_colors.get(action, 'black')
    label = action_labels.get(action, f'Action {action}')
    plt.plot(range(n_generations), averages_low[:, i], color=color, label=label)

plt.title('Low Agents')
plt.xlabel('Period')
plt.ylabel('Choice Frequency')
plt.legend()
plt.show()
```
